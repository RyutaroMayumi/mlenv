```python
import cv2
import numpy as np
from IPython.display import Image, display
```

```python
def imshow(img):
    """ndarray 配列をインラインで Notebook 上に表示する。
    """
    ret, encoded = cv2.imencode(".jpg", img)
    display(Image(encoded))
```

```python
# 画像を読み込む（gray）
img = cv2.imread("/usr/src/app/data/color/lenna.bmp", cv2.IMREAD_GRAYSCALE)
imshow(img)
print(img.shape)
```

```python
# 画像を読み込む（color）
img_color = cv2.imread("/usr/src/app/data/color/lenna.bmp", cv2.IMREAD_COLOR)
imshow(img_color)
print(img_color.shape)
```

```python
# 微分フィルタ
kernel = np.array([[0, 0, 0], [0, -1, 1], [0, 0, 0]])

# フィルタリングを行う。
dst = cv2.filter2D(img, -1, kernel)
imshow(dst)
```

```python
# 画像の正規化
dst = (img / 255.0).astype(np.float32)
imshow(dst)
```

```python
np.set_printoptions(precision=3)
dst
```

```python
# リサイズ
dst = cv2.resize(img, [64, 64])
imshow(dst)
```

```python
# 反転
dst = cv2.flip(img, 1)
imshow(dst)
```

```python
# 回転
trans = cv2.getRotationMatrix2D((img.shape[0]/2, img.shape[1]/2), 45 , 1.0)
dst = cv2.warpAffine(img, trans, (img.shape[0], img.shape[1]))
imshow(dst)
```

```python
# コントラスト調整
dst = cv2.convertScaleAbs(img, alpha=1.5, beta=10)
imshow(dst)
```
